```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://100senmaru.com/yumei-8kpzhxbp/'
request = requests.get(url)
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
text = soup.find('div', 'entry-content cf').find_all(text=True)
```

```python
text1 = []
for i in text:
    text1.append(i.get_text().split())
```

빈 리스트 제거

```python
text2 = list(filter(None, text1))
```

```python
for i in range(len(text2)):
    if len(text2[i][0]) <= 8:
        text2[i].pop()
```

```python
text2 = list(filter(None, text2))
```
