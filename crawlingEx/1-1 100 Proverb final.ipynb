```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://100senmaru.com/yumei-8kpzhxbp/'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

> html 구조가 쉬워보여서 금방 할 줄알았는데, 이건 아니네... 1 ~ 20까지 한 태그, 21 ~ 40까지 한 태그........ ㄱ-
>
> get_text() 했다가 낭패. 별 짓 다해보다가 br 태를 p 태그로 바꿔서 구분짓는걸로... 그러기 위해서 str로 바꿀 수 빡에 없었고.

```python
ptag = list(soup.select('p'))
sp_ptag = []
for i in ptag:
    temp = str(i)
    temp = temp.replace('<br/>', '<p>')
#     for j in temp:
    sp_ptag.append(temp)
' '.join(sp_ptag)
```

```python
import re
```

> 그 결과로 긁어 온 [정규식](https://code-study.tistory.com/55). 왜 긁어 왔는지 참고하셈. 

```python
new = []
for i in sp_ptag:
    new.append(re.sub('(<([^>]+)>)', '', i))
```

> \n 으로 자르기 (리스트화)

```python
new1 = []
for i in range(0, 5+1):
    temp = new[i].split('\n')
    new1.append(temp)
```

```python
new1
```

> 리스트 벗겨내기

```python
li = []
for i in new1:
    if type(i) is list:
        for j in i:
            li.append(j)
    else:
        li.append(i)
```

```python
li
```

> li 의 0 번을 column으로 나머지를 value 로 DataFrame 을 만들고, 중복제거

```python
df = pd.DataFrame(li[1:], columns=[li[0]]).drop_duplicates()
df
```
